```python
import plotly.express as px

df = px.data.stocks()

fig = px.line(df, x='date', y='GOOG')
fig.show()
```

```python
import plotly.express as px

df = px.data.stocks()

fig = px.line(
    df,
    x='date',
    y='GOOG',
    title='Google Stock Price Over Time'
)

fig.show()
```

```python
import plotly.express as px

df = px.data.stocks()

fig = px.line(
    df,
    x='date',
    y=['GOOG', 'MSFT', 'AAPL'],
    title='Stock Prices Comparison'
)

fig.show()
```
