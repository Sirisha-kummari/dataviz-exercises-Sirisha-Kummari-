```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Dataset: CO2 Emissions by Country 2000-2022
# Source: Our World in Data (https://ourworldindata.org/co2-emissions)
df = pd.read_csv('../data/co2_emissions.csv')
print(f"Loaded: {len(df)} rows | Countries: {df['Country'].nunique()} | Years: {df['Year'].min()}-{df['Year'].max()}")
print(df.head())
```

```python
# Explore before building

print("Countries:", df['Country'].unique())
print("\nCO2 range:", df['CO2_Mt'].min(), "to", df['CO2_Mt'].max(), "Mt")
print("\nRegional averages (2022):")
print(df[df['Year']==2022].groupby('Region')['CO2_Mt'].mean().sort_values(ascending=False).round(1))
```

## Task 1 — Multi-Series Line Chart with Highlight

**What to build:** A line chart showing CO2 emissions over time for **all Asian countries** in the dataset, with one country highlighted.

**Requirements:**
- All countries shown (for context), but only **one highlighted in colour** — your choice which
- All other lines in grey (#DDDDDD), thinner
- Highlighted country **labelled directly** at the end of its line (not in a legend)
- Insight title that names the highlighted country and its story

> 💡 `df[df['Region'] == 'Asia']` to filter; use `go.Figure()` with a loop for per-country control


```python
# Task 1 — Multi-series line with highlight
# YOUR CODE HERE
asia = df[df['Region'] == 'Asia']
asia.head()
import plotly.graph_objects as go

fig = go.Figure()

countries = asia['Country'].unique()

highlight_country = 'China'  # you can change this if you want

for country in countries:
    c_data = asia[asia['Country'] == country]

    if country == highlight_country:
        fig.add_trace(go.Scatter(
            x=c_data['Year'],
            y=c_data['CO2_Mt'],
            mode='lines',
            name=country,
            line=dict(color='#E63946', width=3)
        ))
        
        # label at end
        fig.add_annotation(
            x=c_data['Year'].max(),
            y=c_data['CO2_Mt'].values[-1],
            text=country,
            showarrow=True,
            arrowhead=1,
            font=dict(color='#E63946')
        )

    else:
        fig.add_trace(go.Scatter(
            x=c_data['Year'],
            y=c_data['CO2_Mt'],
            mode='lines',
            line=dict(color='#DDDDDD', width=1),
            showlegend=False
        ))

fig.update_layout(
    title="China drives the majority of Asia’s CO2 emissions growth over time",
    xaxis_title="Year",
    yaxis_title="CO2 Emissions (Mt)",
    plot_bgcolor='white',
    showlegend=False
)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=True, gridcolor='#EEEEEE')


fig.show()
```

## Task 2 — Slopegraph: Regional Change 2000 vs 2022

**What to build:** A slopegraph comparing **average regional CO2 emissions** between 2000 and 2022.

**Requirements:**
- One line per region (not per country — aggregate first)
- Colour: regions that increased = one colour; decreased = another
- Values labelled at both ends of each line
- No y-axis tick labels (the endpoint labels make them redundant)
- Insight title stating which regions moved most

> 💡 `df.groupby(['Region','Year'])['CO2_Mt'].mean().reset_index()` then filter to 2000 and 2022


```python
# Task 2 — Slopegraph: regional averages
# YOUR CODE HERE
region_2000_2022 = df.groupby(['Region','Year'])['CO2_Mt'].mean().reset_index()

data_2000 = region_2000_2022[region_2000_2022['Year'] == 2000]
data_2022 = region_2000_2022[region_2000_2022['Year'] == 2022]

slope = pd.merge(
    data_2000[['Region','CO2_Mt']],
    data_2022[['Region','CO2_Mt']],
    on='Region',
    suffixes=('_2000','_2022')
)

slope

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# choose colors based on change
for i in range(len(slope)):
    row = slope.iloc[i]
    
    if row['CO2_Mt_2022'] > row['CO2_Mt_2000']:
        color = '#E63946'  # increased = red
    else:
        color = '#2E75B6'  # decreased = blue
    
    plt.plot([2000, 2022],
             [row['CO2_Mt_2000'], row['CO2_Mt_2022']],
             marker='o',
             color=color)
    
    # label at both ends
    plt.text(1999.5, row['CO2_Mt_2000'], row['Region'], ha='right')
    plt.text(2022.5, row['CO2_Mt_2022'], row['Region'], ha='left')

# clean style
plt.xticks([2000, 2022])
plt.ylabel("Average CO2 Emissions")
plt.title("Asia drives global emissions growth while Europe and North America decline")

plt.grid(axis='y', color='#EEEEEE')
plt.show()
```
